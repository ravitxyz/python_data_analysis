### Floats

```python
# oops! missing values
autos.cylinders.value_counts(dropna=False)
```

```python
# where are they missing?

(
    autos
    [cols]
    .query('cylinders.isna()')
)
```

```python
# chaining - add cylinders and displ columns
(
    autos
    [cols]
    .assign(
        cylinders=autos.cylinders.fillna(0).astype('int8'),
        displ=autos.displ.fillna(0)
    )
    .astype({
        'highway08': 'int8',
        'city08': 'int16',
        'comb08': 'int16',
        'fuelCost08': 'int16',
        'range': 'int16',
        'year': 'int16',
    })
    .describe()
)
```

```python
(
    autos
    [cols]
    .groupby('year')
    .drive
    .nunique()
)
```

```python
# drive and make (in .astype) to category
(
    autos
    [cols]
    .assign(
        cylinders=autos.cylinders.fillna(0).astype('int8'),
        displ=autos.displ.fillna(0).astype('float16'),
        drive=autos.drive.fillna('Other').astype('category')
    )
    .astype({
        'highway08':'int8',
        'city08':'int16',
        'comb08':'int16',
        'fuelCost08': 'int16',
        'range': 'int16',
        'year': 'int16',
        'make': 'category'
    })
    .memory_usage(deep=True)
)
```

```python
# drive and make (in .astype) to category

(
    autos
    [cols]
    .assign(
        cylinders=autos.cylinders.fillna(0).astype('int8'),
        displ=autos.displ.fillna(0).astype('float16'),
        drive=autos.drive.fillna('Other').astype('category')
    )
    .astype({
        'highway08':'int8',
        'city08':'int16',
        'comb08':'int16',
        'fuelCost08': 'int16',
        'range': 'int16',
        'year': 'int16',
        'make': 'category'
    })
    .memory_usage(deep=True)
    .sum()
)
```

```python
# let's inspect trany 
# looks like it has two pieces of information embedded in column
(autos.trany.value_counts(dropna=False))
```

```python
# add new column automatic, speeds from trany, then drop trany
(
    autos
    [cols]
    .assign(
        cylinders=autos.cylinders.fillna(0).astype('int8'),
        displ=autos.displ.fillna(0).astype('float16'),
        drive=autos.drive.fillna('Other').astype('category'),
        automatic=autos.trany.str.contains('Auto'),
        speeds=autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8')
    )
    .astype({
        'highway08':'int8',
        'city08':'int16',
        'comb08':'int16',
        'fuelCost08': 'int16',
        'range': 'int16',
        'year': 'int16',
        'make': 'category'
    })
    .memory_usage(deep=True)
    .sum()
)
```
