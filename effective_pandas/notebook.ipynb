# Types
Getting the right types will enable faster and accureate analysis.

```python
autos[cols].dtypes
```

- int64 special type implemented by numpy for faster operations on large amount of data
- object might be strigns or mixed.
- missing value info can be extracted by the types

```python
autos[cols].memory_usage(deep=True)
```

```python
autos[cols].memory_usage(deep=True).sum()
```

## Ints

```python
# chaining
(
    autos
    [cols]
    .select_dtypes(int)
    .describe()
)
```

- count -- number of non null values

```python
# can comb08 be and int8?
np.iinfo(np.int8)
```

```python
np.iinfo(np.int16)
```

```python
(autos
[cols]
.astype({'highway08':'int8', 'city08': 'int16', 'comb08': 'int16'})
.select_dtypes([int, 'int8'])
.describe()
)
```
